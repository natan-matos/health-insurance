## 2.1 Rename Columns

```python
df1 = df.copy()
```

```python
def rename_cols (data):
    cols_old = ['id', 'Gender', 'Age', 'Driving_License', 'Region_Code',
       'Previously_Insured', 'Vehicle_Age', 'Vehicle_Damage', 'Annual_Premium',
       'Policy_Sales_Channel', 'Vintage', 'Response']
    snake_case = lambda x: inflection.underscore(x)
    cols_new = list(map(snake_case, cols_old))

    # Rename
    data.columns = cols_new

    return data
```

```python
df1 = rename_cols(df1)
df_test = rename_cols(df_test)
```

## 2.2 Feature Engineering

```python
def feature_eng(df1):
    
    # Average Annual premium per region code
    aux1 = df1[['region_code', 'annual_premium']].groupby('region_code').mean().reset_index()

    # Merge with the original dataset
    aux1.rename(columns={'annual_premium': 'mean_premium_by_code'}, inplace=True)
    df1 = pd.merge(df1, aux1, on='region_code', how='left')

    # Insured by code
    df10 = df1[df1['previously_insured'] == 1 ]
    aux1 = df10[['previously_insured', 'region_code']].groupby('region_code').count().reset_index()


    # Merge with the original dataset
    aux1.rename(columns={'previously_insured': 'previously_insured_by_code'}, inplace=True)
    df1 = pd.merge(df1, aux1, on='region_code', how='left')

    # Vahicle Damage
    df1['vehicle_damage'] = df1['vehicle_damage'].apply( lambda x: 1 if x == 'Yes' else 0 )

    # Vahicle Age
    df1['vehicle_age'] = df1['vehicle_age'].apply( lambda x: '1-2' if x == '1-2 Year' else '<1' if x == '> 2 Years' else '>2' )

    # Gender
    df1['gender'] = df1['gender'].apply( lambda x: 0 if x == 'Male' else 1 )

    return df1
```

```python
df1 = feature_eng(df1)
```

```python
df_test = feature_eng(df_test)
```

### Standardization, Rescaling and Encoding

```python
def data_preparation(data):
    
    # STANDARDIZATION
    # Define the scaler
    ss = StandardScaler()

    # Annual Premium
    data['annual_premium'] = ss.fit_transform(data[['annual_premium']].values) 

    # Mean premium by code
    data['mean_premium_by_code'] =  ss.fit_transform(data[['mean_premium_by_code']].values)

    # RESCALING
    mm = MinMaxScaler()

    # Age
    data['age'] = mm.fit_transform(data[['age']].values)

    # vintage
    data['vintage'] = mm.fit_transform(data[['vintage']].values)

    # previously_insured_by_code
    data['previously_insured_by_code'] = mm.fit_transform(data[['previously_insured_by_code']].values)

    # ENCODING
    # Region code
    t_encoder_region = data.groupby('region_code')['response'].mean()
    data.loc[:, 'region_code'] = data['region_code'].map(t_encoder_region)


    # vehicle age
    data = pd.get_dummies(data, prefix='vehicle_age', columns=['vehicle_age'])
    data['vehicle_age_1-2'] = data['vehicle_age_1-2'].astype(int) 
    data['vehicle_age_<1'] = data['vehicle_age_<1'].astype(int)
    data['vehicle_age_>2'] = data['vehicle_age_>2'].astype(int)

    # policy sales channel
    frequency_channel = data.groupby('policy_sales_channel').size() / len(data)
    data.loc[:, 'policy_sales_channel'] = data['policy_sales_channel'].map(frequency_channel)

    return data
```

## 7.3 Performance Comparation

```python
# Merge response column 
df8 = X_validation.copy()
```

### Precision @ K

```python
def precision (data, model, k=20000):
    

    # Merge response columns
    data['response'] = y_validation.copy()

    # Propensity score
    data['score'] = model[:, 1].tolist()

    # sort clients
    data = data.sort_values('score', ascending=False)

    # reset index
    data = data.reset_index(drop=True)

    # create ranking order
    data['rank'] = data.index + 1

    # calculate precision @k
    data['precision_at_k'] = data['response'].cumsum() / data['rank']

    return data.loc[k, 'precision_at_k']
```

### Recall @ K

```python
def recall (data, model, k=20000):

    # Merge response columns
    data['response'] = y_validation.copy()

    # Propensity score
    data['score'] = model[:, 1].tolist()

    # sort clients
    data = data.sort_values('score', ascending=False)

    # reset index
    data = data.reset_index(drop=True)

    # create ranking order
    data['rank'] = data.index + 1

    data['recall_at_k'] = data['response'].cumsum() / data['response'].sum()

    return data.loc[k, 'recall_at_k']
```
